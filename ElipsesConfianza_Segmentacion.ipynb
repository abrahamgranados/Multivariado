**a)**

* Como $\Sigma$ es positiva definida entonces todos sus valores propios 
  $\lambda_i$ son positivos $\Rightarrow |\Sigma| = \prod_{i=1}^{p}     \lambda_i > 0$ 

  $\Rightarrow |\Sigma|^{-\frac{1}{2}} > 0$ y como $\frac{1}{[\sqrt{2\pi}]^{p}} > 0 \Rightarrow  \frac{|\Sigma|^{-\frac{1}{2}}}{[\sqrt{2\pi}]^{p}} > 0$ 

* Para toda $t \in \mathbb{R}$, $exp(t) > 0$, en particular, como $-\frac{1}{2}[\mathbf{x} - \mu]'\Sigma^{-1}[\mathbf{x} - \mu] \in \mathbb{R}$ 

  $\Rightarrow  exp\Big(-\frac{1}{2}[\mathbf{x} - \mu]'\Sigma^{-1}[\mathbf{x} - \mu]\Big) > 0$

Así, multiplicando los dos factores anteriores, tenemos que para toda $\mathbf{x} \in \mathbb{R}^{p}$

$f_{X_1,...,X_p}(x_1,...,x_p) = \frac{|\Sigma|^{-\frac{1}{2}}}{[\sqrt{2\pi}]^{p}}exp\Big(-\frac{1}{2}[\mathbf{x} - \mu]'\Sigma^{-1}[\mathbf{x} - \mu]\Big) > 0$

**b)**

Como $\Sigma$ es positiva definida entonces es simétrica, por lo que podemos ortogonalizarla diagonalmente, i.e. $\Sigma = Q'DQ$ con $D = diag(\lambda_1,...,\lambda_p)$.

Vemos que una raiz cuadrada de la matriz $\Sigma$ es $\Sigma^{\frac{1}{2}} = Q'D^{\frac{1}{2}}Q$ con $D^{\frac{1}{2}} = diag(\sqrt{\lambda_1},...,\sqrt{\lambda_p})$

Observamos que 

$$\Sigma^{\frac{1}{2}}\Sigma^{-1}\Sigma^{\frac{1}{2}}=(Q'D^{\frac{1}{2}}Q)(Q'D^{-1}Q)(Q'D^{\frac{1}{2}}Q)=Q'D^{\frac{1}{2}}D^{-1}D^{\frac{1}{2}}Q=Q'Q=I$$

Mediante el cambio de variable 

$\mathbf{x} =  \Sigma^{\frac{1}{2}}\mathbf{z} + \mu$ 

$|J| = |\Sigma^{\frac{1}{2}}| = |\Sigma|^{\frac{1}{2}}$

$$ = \int_{\mathbb{R}^{p}} exp\Big(-\frac{1}{2}[\mathbf{x} - \mu]'\Sigma^{-1}[\mathbf{x} - \mu]\Big) d\mathbf{z}$$

$$ = \int_{\mathbb{R}^{p}} exp\Big(-\frac{1}{2}\mathbf{z}'\Sigma^{\frac{1}{2}}\Sigma^{-1}\Sigma^{\frac{1}{2}}\mathbf{z}\Big) |\Sigma|^{\frac{1}{2}} d\mathbf{z}$$

$$ = \int_{\mathbb{R}^{p}} exp\Big(-\frac{1}{2}\mathbf{z}'\mathbf{z}\Big) |\Sigma|^{\frac{1}{2}} d\mathbf{z}$$

$$ = |\Sigma|^{\frac{1}{2}} \int_{\mathbb{R}^{p}} exp\Big(-\frac{z_1^2}{2} -\frac{z_2^2}{2} -\cdots-\frac{z_p^2}{2}\Big) d\mathbf{z}$$

$$ = |\Sigma|^{\frac{1}{2}} \int_{\mathbb{R}^{p}} exp\Big(-\frac{z_1^2}{2} -\frac{z_2^2}{2} -\cdots-\frac{z_p^2}{2}\Big) d\mathbf{z}$$

$$ = |\Sigma|^{\frac{1}{2}} \int_{\mathbb{R}^{p}} exp\Big(-\frac{z_1^2}{2}\Big) exp\Big(-\frac{z_2^2}{2}\Big) \cdots exp\Big(-\frac{z_p^2}{2}\Big) d\mathbf{z}$$

$$ = |\Sigma|^{\frac{1}{2}} 
\int_{-\infty}^{\infty}exp\Big(-\frac{z_1^2}{2}\Big)dz_1 \int_{-\infty}^{\infty}exp\Big(-\frac{z_2^2}{2}\Big)dz_2 \cdots 
\int_{-\infty}^{\infty}exp\Big(-\frac{z_p^2}{2}\Big) dz_p $$ 

$$ = |\Sigma|^{\frac{1}{2}} [\sqrt{2\pi}]^{p} $$

pues $\int_{-\infty}^{\infty}exp\Big(-\frac{t^2}{2}\Big)dt = \sqrt{2\pi}$

Por lo tanto, 

$$\int_{\mathbb{R}^{p}} \frac{|\Sigma|^{-\frac{1}{2}}}{\sqrt{2\pi}}  exp\Big(-\frac{1}{2}[\mathbf{x} - \mu]'\Sigma^{-1}[\mathbf{x} - \mu]\Big) d\mathbf{x} = \frac{|\Sigma|^{-\frac{1}{2}}}{[\sqrt{2\pi}]^{p}} |\Sigma|^{\frac{1}{2}} [\sqrt{2\pi}]^{p} = 1 $$

$
M_{X}(t)=exp(\mu t+\frac{1}{2}t^{T}\Sigma t)

$
M_{Y}(t)=E[exp(Yt)]=E[exp((XA+b)t)]=E[exp(XAt)exp(bt)]=E[exp(XAt)]exp(bt)
$

Veamos que la esperanza de la expresión es la generadora de momentos del vector X evaluada en At

$
M_{Y}(t)=M_{X}(At)exp(bt)=exp(\mu At+\frac{1}{2}(At)^{T}\Sigma At)exp(bt)=
$
$
M_{Y}(t)=exp(\mu At+bt+\frac{1}{2}t^{T}A^{T}\Sigma At)=exp((\mu A+b)t+\frac{1}{2}t^{T}A^{T}\Sigma At)
$

La función generadora de momentos de Y, coincide con la f.g.m de una
